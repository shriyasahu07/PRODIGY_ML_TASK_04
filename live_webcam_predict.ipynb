```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model = load_model('gesture_model.h5')
IMG_SIZE = 64
LABELS = ['01_palm', '02_l', '03_fist', '04_fist_moved', '05_thumb', '06_index',
          '07_ok', '08_palm_moved', '09_c', '10_down']

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Crop center square
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    center_crop = cv2.resize(gray, (IMG_SIZE, IMG_SIZE))
    norm_img = center_crop.astype('float32') / 255.0
    norm_img = np.expand_dims(norm_img, axis=(0, -1))

    pred = model.predict(norm_img)
    label = LABELS[np.argmax(pred)]

    cv2.putText(frame, f"Gesture: {label}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
    cv2.imshow("Gesture Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
